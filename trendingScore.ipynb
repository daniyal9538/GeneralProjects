```
#score =  s0*o^0 + s1*o^1 .... + s7*o^6
#sn = alpha(n) + beta (l) + gamma (l1)
# replace l with ni, where ni is n in terms of l, set alpha  = 1
# sn = alpha(n) + ni + gamma(l1)
# since i do not know how l1 will be accessed, i will ignore it for now
#sn = alpha(n) + ni
# ni = c(l) + b
#alpha = 1
#c=0.04566062
#b=0.99779814



#in this script i will get the the score for 1 restaurant




import random 
# create mock data for the number of posts for that restaurant gotten for 8 consequtive days
numPosts = random.sample(range(1, 10), 8)
# create mock data for the number of likes gotten on posts for the restaurant for 8 consequtive days
numLikes = random.sample(range(1, 30), 8)


#hyper parameter values
c=0.04566062
b=0.99779814
alpha = 1
o = 0.85


def getRatingForDay(postCount, likeCount):
    #this function should be called everyday
    rating  = float(alpha*(postCount) + c*(likeCount) + b)
    return rating


def getPostCount(day, restID):
    #use this function to get the post count for restaurant on day
    #only 1 restaurant in this script but for implementation, restaurant should be identified by ID
    return numPosts[day]

def getLikeCount(day, restID):
    #use this function to get the post count for restaurant on day
    #only 1 restaurant in this script but for implementation, restaurant should be identified by ID
    return numLikes[day]

def initRatingDict():
    #use this function to initialize rating array of all restaurants in system
    #the rating array will contain the ratings for each day for the past 7 days for that restaurant
    ratingArr = [0]*7
    #in python the rating arrays will be in dictionray, where key is rest ID, and value is the rating array
    ratingDict ={}
    #all restaurant IDs can be stored in 1 array
    restID = [1]
    for i in restID:
        ratingDict[i]=ratingArr

    return ratingDict

def addNewRating(ratingArr, newRating):
    temp = ratingArr
    temp = [newRating]+temp
    temp.pop()
    ratingArr = temp
    return ratingArr

def getScore(ratingArr):
    score = float(0)
    for i in range(7):
        score += (o**i)*ratingArr[i]
    
    return score

def updateRating(ratingDict, day):
    #function should be called everyday
    for key, value in ratingDict.items():
        likes = getLikeCount(day, key)
        posts = getPostCount(day, key)
        #print(likes, posts)
        rating = getRatingForDay(posts, likes)
        #print(rating)
        ratingArr = addNewRating(value, rating)
        ratingDict[key] = ratingArr
        #print(key, value, day, len(value))

    print(ratingDict)

def main():
    ratingDict = initRatingDict()
    print(numLikes, numPosts)
    for i in range(8):
        
        updateRating(ratingDict, i)
        print(getScore(ratingDict[1]))

if __name__ == '__main__':
    main()
```
